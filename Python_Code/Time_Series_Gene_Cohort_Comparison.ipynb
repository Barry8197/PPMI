# Comparison of Results from Differential Gene Expression Analysis 

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
BL_gPD = pd.read_csv('../outputData/gPD_DEG_BL.csv' , index_col=0)
V02_gPD = pd.read_csv('../outputData/gPD_DEG_V02.csv' , index_col=0)
V04_gPD = pd.read_csv('../outputData/gPD_DEG_V04.csv' , index_col=0)
V06_gPD = pd.read_csv('../outputData/gPD_DEG_V06.csv' , index_col=0)
V08_gPD = pd.read_csv('../outputData/gPD_DEG_V08.csv' , index_col=0)
```

```python
BL_iPD = pd.read_csv('../outputData/iPD_DEG_BL.csv' , index_col=0)
V02_iPD = pd.read_csv('../outputData/iPD_DEG_V02.csv' , index_col=0)
V04_iPD = pd.read_csv('../outputData/iPD_DEG_V04.csv' , index_col=0)
V06_iPD = pd.read_csv('../outputData/iPD_DEG_V06.csv' , index_col=0)
V08_iPD = pd.read_csv('../outputData/iPD_DEG_V08.csv' , index_col=0)
```

```python
de_count_gPD = []
de_count_iPD = []
de_count_overlap = []
for time in ['BL' , 'V02' , 'V04' , 'V06' , 'V08'] : 
    df = pd.read_csv('../outputData/gPD_DEG_'+time+'.csv' , index_col=0)
    df2 = pd.read_csv('../outputData/iPD_DEG_'+time+'.csv' , index_col=0)
    de_count_gPD.append(sum(df['de']))
    de_count_iPD.append(sum(df2['de']))
    de_count_overlap.append(sum(df['de'] & df2['de']))
```

```python
plt.plot(['BL' , 'V02' , 'V04' , 'V06' , 'V08']  , de_count_gPD , 'r.-' , linewidth = 0.5 , label = 'gPD DEGs')
plt.plot(['BL' , 'V02' , 'V04' , 'V06' , 'V08']  , de_count_iPD , 'b.-' , linewidth = 0.5 , label = 'iPD DEGs')
plt.plot(['BL' , 'V02' , 'V04' , 'V06' , 'V08']  , de_count_overlap , 'g.-' , linewidth = 0.5 , label = 'Common DEGs')
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
plt.xlabel('Time point')
plt.ylabel('Number of DEG')
plt.grid()
plt.title('DEG Experiment Comparisons')
plt.savefig('./../Plots/DEG_comparison.png' , bbox_inches = 'tight' , dpi = 200)
```

```python
plt.plot(['BL' , 'V02' , 'V04' , 'V06' , 'V08']  , [6,0,72,20,60] , 'k.-' , linewidth = 1)
plt.xlabel('Time point')
plt.ylabel('Percentage Overlap')
plt.grid()
plt.title('iPD DEG overlap with top 50 gPD DEGs')
plt.savefig('./../Plots/iPD_DEG_overlap.png' , bbox_inches = 'tight' , dpi = 200)
```

## Similarity Comparison of Results

### Sorted by Log Fold Change

```python
ordered_BL_iPD = BL_diff_iPD.reindex(BL_diff_iPD['log2FoldChange'].abs().sort_values(ascending=False).index)
ordered_V02_iPD = V02_diff_iPD.reindex(V02_diff_iPD['log2FoldChange'].abs().sort_values(ascending=False).index)
ordered_V04_iPD = V04_diff_iPD.reindex(V04_diff_iPD['log2FoldChange'].abs().sort_values(ascending=False).index)
ordered_V06_iPD = V06_diff_iPD.reindex(V06_diff_iPD['log2FoldChange'].abs().sort_values(ascending=False).index)
ordered_V08_iPD = V08_diff_iPD.reindex(V08_diff_iPD['log2FoldChange'].abs().sort_values(ascending=False).index)
```

```python
ordered_BL_gPD = BL_diff_gPD.reindex(BL_diff_gPD['log2FoldChange'].abs().sort_values(ascending=False).index)
ordered_V02_gPD = V02_diff_gPD.reindex(V02_diff_gPD['log2FoldChange'].abs().sort_values(ascending=False).index)
ordered_V04_gPD = V04_diff_gPD.reindex(V04_diff_gPD['log2FoldChange'].abs().sort_values(ascending=False).index)
ordered_V06_gPD = V06_diff_gPD.reindex(V06_diff_gPD['log2FoldChange'].abs().sort_values(ascending=False).index)
ordered_V08_gPD = V08_diff_gPD.reindex(V08_diff_gPD['log2FoldChange'].abs().sort_values(ascending=False).index)
```

```python
count_BL =[]
count_V02 = []
count_V04 = []
count_V06 = []
count_V08 = []

batch = 1000
for i in range(1 , len(ordered_BL_gPD)-batch) : 
    count_BL.append(len(set(ordered_BL_gPD.index[i:i+batch]) & set(ordered_BL_iPD.index[i:i+batch]))*100/batch)
    count_V02.append(len(set(ordered_V02_gPD.index[i:i+batch]) & set(ordered_V02_iPD.index[i:i+batch]))*100/batch)
    count_V04.append(len(set(ordered_V04_gPD.index[i:i+batch]) & set(ordered_V04_iPD.index[i:i+batch]))*100/batch)
    count_V06.append(len(set(ordered_V06_gPD.index[i:i+batch]) & set(ordered_V06_iPD.index[i:i+batch]))*100/batch)
    count_V08.append(len(set(ordered_V08_gPD.index[i:i+batch]) & set(ordered_V08_iPD.index[i:i+batch]))*100/batch)
    
 
plt.plot(count_BL , 'r' , label = 'BL')
plt.plot(count_V02 , 'b' , label = 'V02')
plt.plot(count_V04 , 'g' , label = 'V04')
plt.plot(count_V06 , 'y' , label = 'V06')
plt.plot(count_V08 , 'k' , label = 'V08')
plt.title('Similarty of genes absolute log fold change for gPD')
plt.ylabel('Percentage Overlap (%)')
plt.xlabel('Index')
plt.legend()
plt.grid()
plt.savefig('./../Plots/iPD_similariy_logfc.png')
```
